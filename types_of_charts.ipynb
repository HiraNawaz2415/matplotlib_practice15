<h1>Bar Chart 📊</h1>
->A bar chart is used to compare different categories. Each bar represents a category, and its height shows the value.

<h4>Example:</h4>
Imagine you are comparing the number of apples, bananas, cherries, and dates you have.
✅ Use it when you want to compare different groups.

<b> bar()  is used.</b>
Basic Bar Chart	plt.bar(x, y)	plt.bar(["A", "B"], [3, 5])
Horizontal Bars	plt.barh(x, y)	plt.barh(["A", "B"], [3, 5])
Change Color	color="red"	plt.bar(x, y, color="hotpink")
Custom Width	width=0.5	plt.bar(x, y, width=0.1)
Custom Height	height=0.5	plt.barh(x, y, height=0.1)

<h3>Note:</h3>
plt.bar() (vertical bars) → uses width
plt.barh() (horizontal bars) → uses height

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']  # Different categories
values = [5, 7, 3, 8]  # Values for each category

plt.bar(categories, values, color='b', width=0.5)  # Creating a bar chart
plt.xlabel("Categories")  # Label for the x-axis
plt.ylabel("Values")  # Label for the y-axis
plt.title("Bar Chart Example")  # Chart title
plt.show()  # Show the chart
```

```python
# Horizontal bar chart
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])  # Categories
y = np.array([3, 8, 1, 10])  # Values

plt.barh(x, y, height=0.4, color="pink")  # Create a bar chart
plt.show()  # Show the chart
```


<h1>Scatter Plot</h1>
A scatter plot shows how two sets of numbers are related. It helps find patterns, like whether one thing increases when another increases.
use the scatter() function to draw a scatter plot.

The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:

<b>Example:</b>
Think of a scatter plot as plotting random points on a graph to see if there's a trend.
✅ Use it when you want to see the relationship between two things.

```python
import matplotlib.pyplot as plt
import numpy as np  # NumPy helps generate random numbers

x = np.random.rand(50)  # 50 random numbers for X-axis
y = np.random.rand(50)  # 50 random numbers for Y-axis

plt.scatter(x, y, color='green')  # Creating a scatter plot
plt.xlabel("X-axis")  # Label for the x-axis
plt.ylabel("Y-axis")  # Label for the y-axis
plt.title("Scatter Plot Example",c="r")  # Chart title
plt.show()  # Show the chart
```

<h4>Assigning Different Colors to Each Dot</h4>
Instead of giving one color to all dots, we can give each dot a different color.
The c argument is used instead of color, as color does not accept multiple values.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data points
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# Assign a different color to each dot
colors = np.array(["red", "green", "blue", "yellow", "pink", "black",
                   "orange", "purple", "beige", "brown", "gray", "cyan", "magenta"])

plt.scatter(x, y, c=colors)  # Use the c argument for colors
plt.show()
```

<h4>Using Colormaps for Gradient Colors</h4>
A colormap is a gradient of colors that range from one color to another.
The color bar is added to help understand what each color represents.
The darker colors show lower values, and the lighter colors show higher values.
sized the dot using <b>s</b>
Alpha
You can adjust the transparency of the dots with the alpha argument.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data points
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# Color values
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, c=colors, cmap='viridis',s=sizes,alpha=0.5)  # Apply colormap

plt.colorbar()  # Add color bar for reference
plt.show()
```

<h1>Histogram 📈</h1>
A histogram shows the frequency (how often something happens) of a range of values. It’s useful for understanding how data is distributed.

Example:
If you roll a dice 1,000 times, a histogram can show how often each number appears.
✅ Use it when you want to see how data is spread out.

<h3>What the Histogram Shows</h3>
The x-axis represents different values from the dataset.
The y-axis represents the frequency (how many times each value appears).
Taller bars mean that certain values appear more frequently.

<h3>Why Use Bins in a Histogram?</h3>
Imagine you have 1000 numbers ranging from -5 to 5. Instead of showing each number separately, a histogram groups similar values together into bars. Each bin represents a range of values, and the height of the bar shows how many numbers fall into that range.
Using bins helps control the level of detail in the histogram:

Too few bins → Oversimplifies the data.
Too many bins → Overcomplicates the data.
Just right → Clearly shows the data distribution.

Think of a histogram like sorting candies into jars based on color.

If you have 1000 candies, you don’t count each one separately.
Instead, you group similar colors together into different jars.
In a histogram, the candies are like numbers, and the jars are called bins. Each bin groups numbers into a range and counts how many numbers fall into that range.
Without bins: You would see one bar for each number (1, 2, 3, ... 10).
👉 With bins (e.g., 2 bins): The numbers 1-5 go into one bin, and 6-10 go into another bin.
Easy-to-Understand Example
Think of 100 students taking a test. They score between 0 and 100.

👉 If we use 10 bins, each bin covers a 10-point range (0-10, 11-20, 21-30, ...).
👉 If we use 5 bins, each bin covers a 20-point range (0-20, 21-40, ...).
👉 If we use 50 bins, each bin covers only 2 points (0-2, 3-4, ...).

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)  # Generate 1000 random values

plt.hist(data, bins=10, color='purple', alpha=0.7)  # Create histogram
plt.xlabel("Value")  # Label for x-axis
plt.ylabel("Frequency")  # Label for y-axis
plt.title("Histogram Example")  # Chart title
plt.show()  # Show the chart
```

```python
<h1>Pie Chart 🥧<h1>
A pie chart shows parts of a whole. Each slice represents a percentage of the total.

Example:
If you have a basket of fruits, a pie chart can show how much of each fruit you have.


✅ Use it when you want to show proportions or percentages.

<h3>When to Use a Pie Chart?</h3>
✅ Use a pie chart when you want to show proportions or percentages 📊
🚫 Don't use it if you have too many categories (it becomes hard to read).


Each slice (piece) of the pie represents a portion of the total.

<h2>👉 Example:</h2>
Imagine you have a basket of fruits with:

30 apples 🍏
20 bananas 🍌
40 cherries 🍒
10 dates 🌴
A pie chart will show how much of each fruit you have compared to the total.

```

```python

import matplotlib.pyplot as plt

labels = ['Apples', 'Bananas', 'Cherries', 'Dates']  # Categories
sizes = [30, 20, 40, 10]  # Percentages of each category

plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['c', 'yellow', 'green', 'brown'])  
plt.title("Pie Chart Example")  # Chart title
plt.show()  # Show the chart
```

✅ Creates the pie chart with:

sizes: Defines how big each slice is.
labels: Names each slice.
autopct='%1.1f%%': Displays percentages (e.g., "30.0%").
colors: Assigns a color to each slice.

<h2>Start Angle: Change Where the First Slice Starts</h2>
By default, the first slice starts at 0 degrees (pointing to the right).
You can change this using startangle to rotate the chart!

✅ Example: Start the first slice at 90 degrees (pointing up)

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])  # Percentages
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]  # Labels

plt.pie(y, labels=mylabels, startangle=90)  # Rotate chart
plt.show()
```

<h2>Explode:</h2> Make a Slice Stand Out!
What if you want to highlight a certain part of the pie chart?
The explode parameter pulls a slice away from the center.

✅ Example: Make "Apples" slice stand out

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]  # Move "Apples" slice out

plt.pie(y, labels=mylabels, explode=myexplode)  # Apply explode effect
plt.show()
```

<h4> Explanation:</h4>

0.2 = The Apples slice moves outward
0 = The other slices stay in place

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]  # Explode Apples slice

plt.pie(y, labels=mylabels, explode=myexplode, shadow=True)  # Add shadow
plt.show()
```

 <h1>Multiple Plots in One Figure (Using subplots)</h1>
Sometimes, we need to show multiple charts together in the same figure.
We can use plt.subplots() to create a grid of plots (also called "subplots").

<h2>How It Works:</h2>

fig, axs = plt.subplots(2, 2)
👉 This creates a 2×2 grid of plots (4 plots in total).

axs[0, 0] → Top-left plot
axs[0, 1] → Top-right plot
axs[1, 0] → Bottom-left plot
axs[1, 1] → Bottom-right plot

<h1>Example: 4 Different Charts in One Figure</h1>

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
x = np.linspace(0, 10, 100)  # 100 values from 0 to 10
y = np.sin(x)  # Sine function
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
data = np.random.randn(1000)  # Random values for histogram

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)

# Plot 1: Line plot (Top-Left)
axs[0, 0].plot(x, y, 'r')  # 'r' means red line
axs[0, 0].set_title("Line Plot")

# Plot 2: Bar chart (Top-Right)
axs[0, 1].bar(categories, values)  
axs[0, 1].set_title("Bar Chart")

# Plot 3: Scatter plot (Bottom-Left)
axs[1, 0].scatter(x, y, color='blue')  
axs[1, 0].set_title("Scatter Plot")

# Plot 4: Histogram (Bottom-Right)
axs[1, 1].hist(data, bins=20, color='green')  
axs[1, 1].set_title("Histogram")

# Adjust spacing to prevent overlapping
plt.tight_layout()

# Show the figure
plt.show()
```

Breaking Down the Code:
Chart Type	Code	What It Does
Line Plot	axs[0, 0].plot(x, y, 'r')	Draws a red line
Bar Chart	axs[0, 1].bar(categories, values)	Shows bars for categories
Scatter Plot	axs[1, 0].scatter(x, y, color='blue')	Plots points instead of a line
Histogram	axs[1, 1].hist(data, bins=20, color='green')	Groups data into bins
🎯 Key Takeaways:
✅ plt.subplots(rows, cols) → Creates a grid of plots
✅ Each subplot is accessed using axs[row, col]
✅ plt.tight_layout() → Prevents overlapping
